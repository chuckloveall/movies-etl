```python
import pandas as pd
```

```python
oscar_csv = "Resources/the_oscar_award.csv"
oscar_df= pd.read_csv(oscar_csv)
```

```python
oscar_df.columns
```

```python
oscar_renamed = oscar_df.rename(columns={'year_film': 'O_Year',
                                         'category': 'O_Category',
                                         'name': 'O_Nominee',
                                         'film': 'O_Film Title',
                                         'winner': 'O_Win'})
oscar_renamed
```

```python
oscar_reduced = oscar_renamed[['O_Year', 'O_Category', 'O_Nominee', 'O_Film Title', 'O_Win']]
oscar_reduced
```

```python
oscar_reduced.dtypes
```

```python
oscar_year = oscar_reduced.loc[(oscar_reduced['O_Year'] >= 1970)]
oscar_year
```

```python
oscar_year = oscar_year.dropna(how='any')
oscar_year
```

```python
oscar_year.insert(0, 'Award Ceremony', 'Oscars')
oscar_year
```

```python
oscar_year.to_csv('Output/oscar_cleaned.csv', index=False, header=True)
```
