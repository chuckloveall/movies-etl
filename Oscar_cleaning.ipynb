```python
import pandas as pd
```

```python
#load each csv file from resources directory
oscar_csv = "Resources/the_oscar_award.csv"
oscar_df= pd.read_csv(oscar_csv)
oscar_df.head()
```

```python
oscar_df.count()
```

```python
oscar_cleaned_df = oscar_df.dropna(how='any')
oscar_cleaned_df.count()
```

```python
oscar_two_df = oscar_cleaned_df[['year_film', 'category', 'name', 'film', 'winner']]

oscar_two_df.head()
```

```python
oscar_renamed = oscar_two_df.rename(columns={'year_film': 'Year',
                                             'category': 'Award Category',
                                             'name':'Nominee Name',
                                             'film' : 'Film',
                                             'winner': 'Win'})
oscar_renamed.head()
```

```python
oscar_renamed.dtypes
```

```python
reduced_o = oscar_renamed[oscar_renamed['Year'] >= 1970]
reduced_o
```

```python
reduced_o.insert(0, 'Award Ceremony', 'Oscars')
reduced_o
```
