```python
import pandas as pd
```

```python
imdb_movies = 'Resources/IMDb movies.csv'
```

```python
movies_df = pd.read_csv(imdb_movies)
```

movies_df.head()

```python
movies_df
```

```python
movies = movies_df.dropna(how="any")
```

```python
movies.head()
```

```python
movies.dtypes
```

```python
movies["year"] = pd.to_numeric(movies["year"])
```

```python
movies = movies.loc[movies["year"] >= 1970]
```

```python
movies
```

```python
movies.sort_values(by=['year'])
movies
```

```python
movies.to_csv('Output/imbd_movies_cleaned.csv', index=False, header=True)
```
