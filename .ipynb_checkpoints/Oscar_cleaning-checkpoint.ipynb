```python
import pandas as pd
```

```python
#load each csv file from resources directory
oscar_csv = "Resources/the_oscar_award.csv"
oscar_df= pd.read_csv(oscar_csv)
oscar_df.head()
```

```python
reduced = oscar_df[['year_film', 'category', 'name', 'film', 'winner']]
reduced
```

```python
dropped = reduced.dropna(how='any')
dropped
```

```python
renamed = dropped.rename(columns={
                            'year_film': 'Year',
                            'category': 'Award Category',
                            'name': 'Nominee Name',
                            'film': 'Film',
                            'winner': 'Win'
})
renamed
```

```python
renamed.dtypes
```

```python
year = renamed[renamed['Year'] >= 1970]
year
```

```python
year.insert(0, 'Award Ceremony', 'Oscars')
year
```

```python
oscar_final = year
oscar_final
```
